```python
import os, re, smtplib
from email.message import EmailMessage

def load_env(path='.env'):
    env = {}
    with open(path, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            if not line or line.startswith('#'):
                continue
            if '=' not in line:
                continue
            key, value = line.split('=', 1)
            key = key.strip()
            value = value.strip()
            if value.startswith('
'):
                value = value[1:-1]\n
            env[key] = value
    return env

env = load_env('c:/Users/ADMIN/Desktop/code/cactistock/.env')
required = ['SMTP_HOST', 'SMTP_PORT', 'SMTP_USER', 'SMTP_PASS', 'SMTP_FROM']
missing = [k for k in required if not env.get(k)]
print('Missing:', missing)
if missing:
    raise SystemExit('Missing SMTP settings')

msg = EmailMessage()
msg['Subject'] = 'CactiStock SMTP test'
msg['From'] = env['SMTP_FROM']
msg['To'] = 'cactistockfiles@gmail.com'
msg.set_content('This is a test email from the local CactiStock app.')

port = int(env['SMTP_PORT']) if env['SMTP_PORT'] else 587
smtp_host = env['SMTP_HOST']
smtp_user = env['SMTP_USER']
smtp_pass = env['SMTP_PASS']
secure = port == 465

print('Sending via', smtp_host, 'port', port, 'secure', secure)
with smtplib.SMTP(smtp_host, port, timeout=15) as server:
    server.starttls() if not secure else None
    server.login(smtp_user, smtp_pass)
    server.send_message(msg)
print('Sent successfully')
```
